```python
with open("The_Bitter_Lesson.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()
print("Total # of characters : ", len(raw_text))
print(raw_text[:99])
```

```python
import re
preprocessed = re.split(r'([.,!;:?_"()\'\-]|--|\s)',raw_text)
preprocessed = [item.strip() for item in preprocessed if item.strip()]
print(len(preprocessed))
print(preprocessed[:37])
```

```python
all_tokens = sorted(list(set(preprocessed)))
all_tokens.extend(["<|endoftext|>", "<|unk|>"])
vocabulary = {token:integer for integer,token in enumerate(all_tokens)}
print(len(vocabulary.items()))
```

```python
for i, item in enumerate(list(vocab.items())[-5:]):
    print(item)
```

```python
class SimpleTokenizerV2:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = { i:s for s,i in vocab.items()}

    def encode(self, text):

        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)

        preprocessed = [
            item.strip() for item in preprocessed if item.strip()
        ]
        preprocessed = [item if item in self.str_to_int
            else "<|unk|>" for item in preprocessed]
        ids = [self.str_to_int[s] for s in preprocessed]
        return ids
    
    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)
        return text
```

```python
text1 = "Hello, do you like tea?"
text2 = "In the sunlit terraces of the palace."
text = " <|endoftext|> ".join((text1, text2))
print(text)
```

```python
tokenizer = SimpleTokenizerV2(vocabulary)
print(tokenizer.encode(text))
```

```python
print(tokenizer.decode(tokenizer.encode(text)))
```

```python
# Byte pair encoding using tiktoken library
# it is a tokenizer
import tiktoken

tokenizer = tiktoken.get_encoding("gpt2")
text = (
"Hello, do you like tea? <|endoftext|> In the sunlit terraces"
"of someunknownPlace."
)
#text2 = "<|endoftext|>"
integers = tokenizer.encode(text, allowed_special={"<|endoftext|>"})
print(integers)

#integers2 = tokenizer.encode(text2, allowed_special={"<|endoftext|>"})
#print(integers2)

```

```python
# exercise 2.1
text = "Akwirw ier"

integers = tokenizer.encode(text, allowed_special={"<|endoftext|>"})
print("Encoding :")
print(integers)

print("\nMapping individual code to token :")

for code in integers:
    temp = [code]
    decode = tokenizer.decode(temp) # this function need list as input
    print(f"{code} <- \"{decode}\"")

print("\nDecode method on integers :")

print(tokenizer.decode(integers))
```
