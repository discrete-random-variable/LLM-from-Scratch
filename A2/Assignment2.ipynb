```python
with open("The_Bitter_Lesson.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()
print("Total # of characters : ", len(raw_text))
print(raw_text[:99])
```

```python
import re
preprocessed = re.split(r'([.,!;:?_"()\'\-]|--|\s)',raw_text)
preprocessed = [item.strip() for item in preprocessed if item.strip()]
print(len(preprocessed))
print(preprocessed[:37])
```

```python
all_words = sorted(set(preprocessed))
vocab_size = len(all_words)
print(vocab_size)
```

```python
# creating vocabulary
vocabulary = {token:integer for integer,token in enumerate(all_words)}
for i,item in enumerate(vocabulary.items()):
    print(item)
    if i >= 37:
        break
```

```python
class SimpleTokenizerV1:
    def __init__(self,vocab):
        self.str_to_int = vocab
        self.int_to_str = {i:s for s,i in vocab.items()}

    def encode(self,text):
        preprocessed = re.split(r'([,.?_!"()\']|--|\s)', text)
        preprocessed = [
        item.strip() for item in preprocessed if item.strip()
        ]
        ids = [self.str_to_int[s] for s in preprocessed]
        return ids

    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)
        return text

tokenizer = SimpleTokenizerV1(vocabulary)
text = """One thing that should be learned from the bitter lesson is the great power of general purpose methods, of methods that continue to scale with increased computation even as the available computation becomes very great. The two methods that seem to scale arbitrarily in this way are search and learning."""
ids = tokenizer.encode(text)
print(ids)
```

```python
print(tokenizer.decode(ids))
```
