```python
import re
text = "Science is elegant. This is exhilarating, get excited!"
result = re.split(r'([.,!]|\s)',text)
print(result)
```

```python
import re
text = "Science is elegant. This is exhilarating -- get excited!"
result = re.split(r'([.,!;:?_"()\']|--|\s)',text)
result = [item.strip() for item in result if item.strip()]
print(result)
```
